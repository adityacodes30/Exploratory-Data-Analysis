```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
colnames = ['vendor', 'Name' ,'MYCT' ,'MMIN' ,'MMAX' ,'CACH' ,'CHMIN','CHMAX','PRP'  ,'ERP'  ]

df = pd.read_csv('machine.data' , names=colnames)

df.head()
```

```python
#checking for null values 
df.isnull()
```

```python
#plotting of heatmap of null values 

sns.heatmap(df.isnull(),yticklabels=False,cbar=False , cmap='viridis')
```

```python
df.dtypes
```

```python
df = df.drop('vendor', axis=1)
df = df.drop('Name', axis=1)
```
